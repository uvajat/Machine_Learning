```python
# BEKFURR INC 2025
# Kutubxonalarni yuklaymiz
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import operator

from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.preprocessing import PolynomialFeatures
```

```python
# EXAMPLE datani olamiz
pl_data = pd.read_csv('/content/drive/MyDrive/Artificial Intelligence/DATA/Polinomial Regression/Position_Salaries.csv')
print(pl_data)
```

```python
# va uni x va y ga ajratamiz
X = pl_data.iloc[:, 1:2].values
Y = pl_data.iloc[:, 2].values
```

```python
# INTERESTING chizib ko'ramiz
```

```python
plt.scatter(X, Y, s=10)
plt.show()
```

```python
#(❁´◡`❁) endi polimial regressiaga bizning datamizni fit qilamiz
# va linear modulni polimial regressia orqali tranformatsiya bilan o'qitamiz
poly = PolynomialFeatures(degree = 2)
X_poly = poly.fit_transform(X)

poly.fit(X_poly, Y)
lin_poly = LinearRegression()
lin_poly.fit(X_poly, Y)
```

```python
#HMMM
#HA xatolik foizini tekshiramiz
```

```python
rmse = np.sqrt(mean_squared_error(Y,lin_poly.predict(poly.fit_transform(X))))
r2 = r2_score(Y,lin_poly.predict(poly.fit_transform(X)))
print(rmse)
print(r2)
```

```python
#Natijani vizuallashtiramiz
```

```python
plt.scatter(X, Y, color = 'blue')

plt.plot(X, lin_poly.predict(poly.fit_transform(X)), color = 'red')
plt.title('Polynomial Regression')
plt.xlabel('Postiton level')
plt.ylabel('Salary')

plt.show()
```
